```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def read_citation_overview(filename):
    df = pd.read_csv(filename, skiprows=5)
    cols = df.columns.values
    cols[:7] = df.iloc[0].values[:7]
    df.columns = cols
    df = df.drop(0)
    return df
```

```python
sns.set_context('talk')

oa_oc = ['cto-oa-oc-authorAZ.csv', 'cto-oa-oc-titleAZ.csv','cto-oa-oc-titleZA.csv']
oa = ['cto-oa-authorAZ.csv', 'cto-oa-titleAZ.csv','cto-oa-titleZA.csv']
oa_arg = ['cto-oa-arg-authorAZ.csv','cto-oa-arg-titleAZ.csv','cto-oa-arg-titleZA.csv']
oa_oc_arg = ['cto-oa-oc-arg.csv']

df_oa_oc = pd.concat([read_citation_overview(xi) for xi in oa_oc])
df_oa_oc = df_oa_oc.drop_duplicates()
df_oa_oc['total'].mean()

df_oa = pd.concat([read_citation_overview(xi) for xi in oa])
df_oa = df_oa.drop_duplicates()
df_oa['total'].mean()

df_oa_oc_arg = pd.concat([read_citation_overview(xi) for xi in oa_oc_arg])
df_oa_oc_arg = df_oa_oc_arg.drop_duplicates()
df_oa_oc_arg['total'].mean()

df_oa_arg = pd.concat([read_citation_overview(xi) for xi in oa_arg])
df_oa_arg = df_oa_arg.drop_duplicates()
df_oa_arg['total'].mean()

years = df_oa_oc.columns[8:-4]
citations_oc = df_oa_oc[df_oa_oc.columns[8:-4]].values.mean(axis=0)
citations_oa = df_oa[df_oa.columns[8:-4]].values.mean(axis=0)
citations_oc_arg = df_oa_oc_arg[df_oa_oc_arg.columns[8:-4]].values.mean(axis=0)
citations_oa_arg = df_oa_arg[df_oa_arg.columns[8:-4]].values.mean(axis=0)

plt.figure(figsize=(10,5))
plt.plot(years,citations_oc_arg,label='Open source Argentina',color=(73./255,119./255,161./255),linestyle='-')
plt.plot(years,citations_oa_arg,label='Open access Argentina',color=(73./255,119./255,161./255),linestyle='--')
plt.plot(years,citations_oc,label='Open source',color=(2./255,202./255,94./255),linestyle='-')
plt.plot(years,citations_oa,label='Open access',color=(2./255,202./255,94./255),linestyle='--')
plt.xlabel('Año')
plt.xticks(ticks=range(1,14,2))
plt.ylabel('Citas por artículo')
plt.grid('on', linestyle='--')
plt.legend()
plt.show()
```
